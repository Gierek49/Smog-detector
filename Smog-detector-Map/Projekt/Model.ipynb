```python
import pandas as pd
import numpy as np

from sklearn.neighbors import KNeighborsClassifier
#from sklearn.model_selection import cross_val_score

%matplotlib inline
```

```python
minLat =49
maxLat=54.50
minLng =14.07
maxLng=24.08
```

```python
api = pd.read_json('http://api.gios.gov.pl/pjp-api/rest/station/findAll')

api.to_csv('Data.csv',index=False)

dft=pd.read_csv('Data.csv')

df=dft[['id', 'stationName', 'gegrLat', 'gegrLon']]

v = df['id'].count()

tid= pd.Series([],dtype=int)

tid=df['id']

Type_new = pd.Series([],dtype=str)

v
```

```python
def take_polution(id):
    return pd.read_json('http://api.gios.gov.pl/pjp-api/rest/aqindex/getIndex/'+str(id) )


for x in range(0,v):
    t = str(tid[x])
    w = take_polution(t)
    Type_new[x]=w['stIndexLevel'].indexLevelName 

```

```python
df.insert(4, "NameLevel", Type_new)
```

```python
Type_new.value_counts()
```

```python
X=df[['gegrLat','gegrLon']]

y=df['NameLevel'] 

n=3

model= KNeighborsClassifier(n_neighbors=n)
model.fit(X,y)
```

```python
def Predict(x,y,model):
    a = np.zeros([1,2], float)
    a[0][0]=y
    a[0][1]=x
    return model.predict(a)
```

```python
def Download_From_API():
    api = pd.read_json('http://api.gios.gov.pl/pjp-api/rest/station/findAll')
    df=api[['id', 'stationName', 'gegrLat', 'gegrLon']]
    return df
```

```python
def Take_Polution(v,tid):
    Type_new = pd.Series([],dtype=str)
    for x in range(0,v):
        t = str(tid[x])
        w = take_polution(t)
        Type_new[x]=w['stIndexLevel'].indexLevelName
    return Type_new
```

```python
def DataToLearn():
    df=Download_From_API()
    v = df['id'].count()
    tid= pd.Series([],dtype=int)
    tid=df['id']
    Type_new = Take_Polution(v,tid)
    df.insert(4, "NameLevel", Type_new)
    return df
    
```

```python
def LearnModel(n):
    df=DataToLearn()
    X=df[['gegrLat','gegrLon']]
    y=df['NameLevel'] 
    model= KNeighborsClassifier(n_neighbors=n)
    model.fit(X,y)
    return model
```

```python
model=LearnModel(3)
```

```python
Predict(53.493550,18.762139,model)
```
